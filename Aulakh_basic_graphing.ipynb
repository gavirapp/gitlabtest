#### Basic Plotting

- We will use a module called `matplotlib` to plot some simple graphs.

- This module provides functions which are very similar to MATLAB plotting commands.

```python
import matplotlib.pyplot as plt
%matplotlib inline

y = x*2 + 5
plt.plot(x, y)
```

```python
# Plotting a sine curve

from numpy import pi, sin

fig = plt.figure(figsize=(10,8))
x = np.arange(0, 6*pi, 0.01)
y = sin(x)
plt.plot(x, y)
plt.show()
```

#### Now let us get some external data and work with it

```python
import pandas as pd
iris = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/iris.csv", index_col=0)
```

```python
iris.tail(8)
```

#### Basic Graphing

```python
#Here is a most basic plot of one of the variables

plt.figure(figsize=(12,8))
plt.plot(iris['Sepal.Length'])
plt.xlabel('ID')
plt.ylabel('Sepal Length')
plt.show()
```

#### <font color= 'blue'> Exercise:  Now it is your turn: Graph of the column "PetalLengthCm" </font>

```python
plt.figure(figsize=(12,8))
plt.plot(iris['Petal.Length'])
plt.xlabel('ID')
plt.ylabel('Petal Length')
plt.show()
```

#### If you want to see many columns graphed in the same plot

```python
# Let us first grab the column names
column_names = [column for column in iris]  #list comprehension
print(column_names)  #check out what we got
print(iris.columns)
print(list(iris.columns))
```

```python
# We are going to use a for loop to plot the numerical variable columns.  Here is a for loop:
for x in range(0, 3):
    print("We're on time %d" % (x))
```

```python
#Here is a way to find the maximum and minimum values the numerical data
mini = iris[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']].min().min() #first min gives min from each column, next one takes min value of those (min value of the entire dataset)
maxi = iris[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']].max().max()
print(mini, maxi)
```

```python
plt.figure(figsize=(16,10))
for i in range(0,4):
        plt.plot(iris[column_names[i]])
plt.ylim(mini,maxi)  #just the plot the range of the data
plt.legend(['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'])  #makes it easier to read
plt.show()
```

### Subplots: Now if we want to see many columns graphed at the same time

```python
plt.figure(figsize=(18,14))
plt.subplot(1,4,1)
plt.plot(iris['Sepal.Length'])
plt.xlabel('Id')
plt.ylabel('Sepal Length')

plt.subplot(1,4,2)
plt.plot(iris['Sepal.Width'],'b')
plt.xlabel('Id')
plt.ylabel('Sepal Width')

plt.subplot(1,4,3)
plt.plot(iris['Petal.Length'],'b')
plt.xlabel('Id')
plt.ylabel('Petal Length')

plt.subplot(1,4,4)
plt.plot(iris['Petal.Width'],'r')
plt.xlabel('Id')
plt.ylabel('Petal Width')

plt.show()
```

#### Slicing Data Frames

```python
#Here we slice the first 100 rows of iris into another dataframe
iris_rows100 = iris.loc[0:99]
iris_rows100.tail()
```

```python
#Here we slice the first 4 columns of iris into data frame
iris1 = iris.loc[:, :'Petal.Width']
iris1.tail()
```

```python

plt.figure(figsize=(12,8))
iris.boxplot(['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']) #this does the job the box plots of all columns
plt.show()
```

#### Here are a bit more advanced graphing:  Regression line with a confidence band

```python
sns.lmplot(data = iris, x = 'Sepal.Length', y = 'Petal.Length')
plt.show()
```

```python
# The same data but now broken into their "species"
sns.lmplot(data = iris, x = 'Sepal.Length', y = 'Petal.Length',hue="Species")
plt.show()
```
